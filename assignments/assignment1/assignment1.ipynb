```python
transactions = [
    {"id": 1, "product": "Laptop", "price": 1200, "quantity": 2, "region": "EU"},
    {"id": 2, "product": "Phone", "price": 800, "quantity": -1, "region": "US"},
    {"id": 3, "product": "Tablet", "price": 0, "quantity": 5, "region": "EU"},
    {"id": 4, "product": "Monitor", "price": 350, "quantity": 3, "region": "US"},
    {"id": 5, "product": "Laptop", "price": 1200, "quantity": 1, "region": "EU"},
    {"id": 6, "product": "Keyboard", "price": 75, "quantity": 10, "region": "APAC"},
]
```

```python
valid_transactions = []
for transaction in transactions:
    if transaction["price"] != 0 and transaction["quantity"] >= 0:
        valid_transactions.append(transaction)
for transaction in valid_transactions:
    transaction["total"] = transaction["price"]*transaction["quantity"]

#grouping all valid transactions per region (dictionaries EU, US and APAC will only have valid transactions):

EU = []
for transaction in valid_transactions:
    if transaction["region"] == "EU":
        EU.append(transaction)
US = []
for transaction in valid_transactions:
    if transaction["region"] == "US":
        US.append(transaction)
APAC = []
for transaction in valid_transactions:
    if transaction["region"] == "APAC":
        APAC.append(transaction)

#making a new variable for each region to store their total revenues (EUtotal, UStotal and APACtotal):

EUtotal = 0
UStotal = 0
APACtotal = 0

for transaction in EU:
    EUtotal = EUtotal + transaction["total"]
    
for transaction in US:
    UStotal = UStotal + transaction["total"]

for transaction in APAC:
    APACtotal = APACtotal + transaction["total"]

#making a dictionary that stores the total revenues for every region:

total_per_region = [
    {"EU": EUtotal},
    {"US": UStotal},
    {"APAC": APACtotal},
]

#comparing the three possible outcomes:

if EUtotal >= UStotal and EUtotal >= APACtotal:
    print(f"The region with the highest revenue is EU with {EUtotal}")
elif UStotal >= EUtotal and UStotal >= APACtotal:
    print(f"The region with the highest revenue is US with {UStotal}")
else:
    print(f"The region with the highest revenue is APAC with {APACtotal}")


```

```python
sales = [1200, 980, 1540, 870, 2010, 1650, 1300, 1100, 950, 1780, 2200, 1430, 1600, 1900]
```

```python
def summarise_sales(list):
    total = (sum(list))
    average = total / len(list)
    best_day = (max(list))
    worst_day = (min(list))
    above_average_count = 0
    
    for sale in list:
        if sale > average:
            above_average_count = above_average_count + 1

    sales_info = [
        {"total": total},
        {"average": average},
        {"best_day": best_day},
        {"worst_day": worst_day},
        {"above_average_count": above_average_count},
    ]
    return sales_info
result = summarise_sales(sales)
print(result)
```

```python
#Task 4:

class Portfolio:
    def __init__(self, owner):
        self.owner = owner
        self.holdings = {}
    def add_asset(self, ticker, shares, price_per_share):
        self.holdings[ticker] = {
            "shares": shares,
            "price_per_share": price_per_share
        }
    def total_value(self):
        total = 0
        for asset in self.holdings.values():
            total = total + asset["shares"] * asset["price_per_share"]
        return total

    def most_valuable(self):
        top_ticker = None
        max_value = 0

        for ticker, asset in self.holdings.items():
            value = asset["shares"] * asset["price_per_share"]
            if value > max_value:
                max_value = value
                top_ticker = ticker

        return top_ticker
    
    def __str__(self):
        return(
            f"Owner: {self.owner}, "
            f"Number of assets: {len(self.holdings)},"
            f"Total value:, {self.total_value()}"
        )

p = Portfolio("User")

p.add_asset("ABC", 40, 100)
p.add_asset("CBA", 33, 22)
p.add_asset("NVDA", 3, 200)

print(p)
print(f"total value of portfolio:  {p.total_value()}")
print(f"most valuable asset: {p.most_valuable()}")

    
        
```

```python
#Task 5:

import datetime
orders = [
    ("2025-01-15", "Laptop"), ("2025-01-22", "Phone"),
    ("2025-02-03", "Laptop"), ("2025-02-14", "Tablet"),
    ("2025-03-01", "Phone"), ("2025-03-10", "Phone"),
    ("2025-03-22", "Laptop"), ("2025-04-05", "Monitor"),
]
dates = []
for date, product in orders:
    #because datetime.date does not work with strings (the format the dates are currently in), 
    #we must first make them integers
    year = int(date[0:4])
    month = int(date[5:7])
    day = int(date[8:10])
    date_integers = datetime.date(year, month, day)
    dates.append(date_integers)

time_between = max(dates) - min(dates)
print(time_between)
```

```python
rng = np.random.default_rng(seed=123)
days = pd.date_range(start="2025-01-06", periods=28, freq="D")
visitors = rng.integers(200, 600, size=28)
conversion_rate = rng.uniform(0.02, 0.08, size=28).round(4)
avg_order_value = rng.normal(50, 10, size=28).round(2)

df = pd.DataFrame({
    "visitors": visitors,
    "conversion_rate": conversion_rate,
    "avg_order_value": avg_order_value
}, index=days)

df.index.name = "date"

df["conversions"] = (df["visitors"] * df["conversion_rate"]).round().astype(int)
df["daily_revenue"] = df["conversions"] * df["avg_order_value"]

df
```
